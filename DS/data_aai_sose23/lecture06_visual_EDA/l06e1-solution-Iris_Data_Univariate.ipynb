Create 4 histograms, one for each of the measured attributes (petal length / width, sepal length / width) and a frequency plot for the target.

```python
%matplotlib inline
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('data/iris_clean.csv', index_col=0)
iris[:10]
```

```python
_ = iris.hist(bins=10)
```

```python
_=iris.hist(bins=20)
```

```python
_ = iris.hist(bins=30, figsize=(12,7), sharex=True, sharey=True)
```

```python
# solution with seaborn incl. density plots. The use of the shared axes allows us to see the histograms
# in relation to each other
f, axes = plt.subplots(2, 2, figsize=(12, 7), sharex=True, sharey=True)
sns.histplot( iris["petal length"] , kde=True, color="blue", bins=20, ax=axes[0, 0])
sns.histplot( iris["petal width"] , kde= True, color="blue", bins=20, ax=axes[0, 1])
sns.histplot( iris["sepal length"] , kde= True, color="red", bins=20, ax=axes[1, 0])
_=sns.histplot( iris["sepal width"] , kde=True, color="red", bins=20, ax=axes[1, 1])
```

```python
# frequency plot 
_=sns.countplot(data=iris, x="class")
```

20 bins seems to be a good number. 
- The histograms for petal length and width look like two normal distributions superimposed, one on the left and one on the right side.
- The histograms for sepal length and width look vaguely like normal distributions. 
- We have 50 instances of each class, the classes are very nicely balanced

```python
bins=25
fig, axs = plt.subplots(2, 2, figsize=(12, 7))

column='petal length'
axs[0, 0].hist(iris[iris["class"] == "Iris-setosa"][column], bins=bins, color="blue")
axs[0, 0].hist(iris[iris["class"] == "Iris-versicolor"][column], bins=bins, color="orange")
axs[0, 0].hist(iris[iris["class"] == "Iris-virginica"][column], bins=bins, color="green")
axs[0, 0].set_title(column)

column='petal width'
axs[0, 1].hist(iris[iris["class"] == "Iris-setosa"][column], bins=bins, color="blue")
axs[0, 1].hist(iris[iris["class"] == "Iris-versicolor"][column], bins=bins, color="orange")
axs[0, 1].hist(iris[iris["class"] == "Iris-virginica"][column], bins=bins, color="green")
axs[0, 1].set_title(column)

column='sepal length'
axs[1, 0].hist(iris[iris["class"] == "Iris-setosa"][column], bins=bins, color="blue")
axs[1, 0].hist(iris[iris["class"] == "Iris-versicolor"][column], bins=bins, color="orange")
axs[1, 0].hist(iris[iris["class"] == "Iris-virginica"][column], bins=bins, color="green")
axs[1, 0].set_title(column)

column='sepal width'
axs[1, 1].hist(iris[iris["class"] == "Iris-setosa"][column], bins=bins, color="blue")
axs[1, 1].hist(iris[iris["class"] == "Iris-versicolor"][column], bins=bins, color="orange")
axs[1, 1].hist(iris[iris["class"] == "Iris-virginica"][column], bins=bins, color="green")
_=axs[1, 1].set_title(column)
```

```python
# solution with seaborn. 
f, axes = plt.subplots(2, 2, figsize=(12, 7))

column='petal length'
sns.histplot( iris[iris["class"] == "Iris-setosa"][column] , kde=True, color="blue", edgecolor=None, bins=20, ax=axes[0, 0])
sns.histplot( iris[iris["class"] == "Iris-versicolor"][column] , kde=True, color="orange", edgecolor=None, bins=20, ax=axes[0, 0])
sns.histplot( iris[iris["class"] == "Iris-virginica"][column] , kde=True, color="green", edgecolor=None, bins=20, ax=axes[0, 0])
column='petal width'
sns.histplot( iris[iris["class"] == "Iris-setosa"][column] , kde=True, color="blue", edgecolor=None, bins=20, ax=axes[0, 1])
sns.histplot( iris[iris["class"] == "Iris-versicolor"][column] , kde=True, color="orange", edgecolor=None, bins=20, ax=axes[0, 1])
sns.histplot( iris[iris["class"] == "Iris-virginica"][column] , kde=True, color="green", edgecolor=None, bins=20, ax=axes[0, 1])
column='sepal length'
sns.histplot( iris[iris["class"] == "Iris-setosa"][column] , kde=True, color="blue", edgecolor=None, bins=20, ax=axes[1, 0])
sns.histplot( iris[iris["class"] == "Iris-versicolor"][column] , kde=True, color="orange", edgecolor=None, bins=20, ax=axes[1, 0])
sns.histplot( iris[iris["class"] == "Iris-virginica"][column] , kde=True, color="green", edgecolor=None, bins=20, ax=axes[1, 0])
column='sepal width'
sns.histplot( iris[iris["class"] == "Iris-setosa"][column] , kde=True, color="blue", edgecolor=None, bins=20, ax=axes[1, 1])
sns.histplot( iris[iris["class"] == "Iris-versicolor"][column] , kde=True, color="orange", edgecolor=None, bins=20, ax=axes[1, 1])
_=sns.histplot( iris[iris["class"] == "Iris-virginica"][column] , kde=True, color="green", edgecolor=None, bins=20, ax=axes[1, 1])
```

So it looks like Iris Setosa is very well seperated on the petal-columns from the other two kinds. Versicolor and Virginica are not clearly seperated. There is no clear separation on the petal attributes at all.
