# Iris Dataset revisited

# Part IV - Visual EDA - Bivariate and Multivariate

IV.1: Create 4 scatter plots, showing
- the petal length vs. the petal width
- the sepal length vs. the sepal width
- the petal length vs. the sepal length
- the petal width vs. the sepal width.

In all scatter plots, mark the type of Iris a data point pertains to by a color (e.g. blue for Iris setosa, red for Iris versicolor, black for Iris virginica).

Document your insights / discoveries.

```python
## Solutions
```

```python
%matplotlib inline
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('data/iris_clean.csv', index_col=0)
iris[:10]
```

```python
## ----- IV.1
```

```python
setosas = iris[iris['class']=='Iris-setosa']
versicolors = iris[iris['class']=='Iris-versicolor']
virginicas = iris[iris['class']=='Iris-virginica']

fig, axis = plt.subplots(2, 2)
axis[0,0].scatter(setosas['petal length'], setosas['petal width'], color='blue', alpha=0.2)
axis[0,0].scatter(versicolors['petal length'], versicolors['petal width'], color='red')
axis[0,0].scatter(virginicas['petal length'], virginicas['petal width'], color='black')
axis[0,1].scatter(setosas['sepal length'], setosas['sepal width'], color='blue')
axis[0,1].scatter(versicolors['sepal length'], versicolors['sepal width'], color='red')
axis[0,1].scatter(virginicas['sepal length'], virginicas['sepal width'], color='black')

axis[1,0].scatter(setosas['petal length'], setosas['sepal length'], color='blue')
axis[1,0].scatter(versicolors['petal length'], versicolors['sepal length'], color='red')
axis[1,0].scatter(virginicas['petal length'], virginicas['sepal length'], color='black')
axis[1,1].scatter(setosas['petal width'], setosas['sepal width'], color='blue')
axis[1,1].scatter(versicolors['petal width'], versicolors['sepal width'], color='red')
axis[1,1].scatter(virginicas['petal width'], virginicas['sepal width'], color='black')
_ = _
```

```python
# nice alternative
ic = {'Iris-setosa':'blue', 'Iris-versicolor':'red', 'Iris-virginica':'black'}

fig, axis = plt.subplots(2, 2)
axis[0,0].scatter(iris['petal length'], iris['petal width'], color=iris['class'].apply(lambda x:ic[x]))
axis[0,1].scatter(iris['sepal length'], iris['sepal width'], color=iris['class'].apply(lambda x:ic[x]))
axis[1,0].scatter(iris['petal length'], iris['sepal length'], color=iris['class'].apply(lambda x:ic[x]))
axis[1,1].scatter(iris['petal width'], iris['sepal width'], color=iris['class'].apply(lambda x:ic[x]))
_ = _
```

```python
# or even simpler
ic = {'Iris-setosa':'blue', 'Iris-versicolor':'red', 'Iris-virginica':'black'}

fig, axis = plt.subplots(2, 2)
axis[0,0].scatter(iris['petal length'], iris['petal width'], color=iris['class'].map(ic))
axis[0,1].scatter(iris['sepal length'], iris['sepal width'], color=iris['class'].map(ic))
axis[1,0].scatter(iris['petal length'], iris['sepal length'], color=iris['class'].map(ic))
axis[1,1].scatter(iris['petal width'], iris['sepal width'], color=iris['class'].map(ic))
_ = _
```

```python
# simple (but time consuming for larger datasets) solution with seaborn
_=sns.pairplot(iris, hue='class')
```

- Iris setosa is clearly separated from the other two in all four scatter plots and also in petal length and petal width: Iris setosa has the smallest petals, and can be identified by this property.
- The other two seem very close together, hard to tell apart in 1-dimesional and 2-dimensional projections, may be possible in higher dimensions.
